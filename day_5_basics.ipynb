*Write a function that takes a list of numbers and returns a new list containing only the numbers greater than 10.*

```python

def numset(s):
    new_list = []
    for i in s:
        if i > 10:
            new_list.append(i)
    return new_list


s=[7,56,6,45,31,13,2,5]
numset(s)
```

*Write a function to check whether two words are anagrams. (Definition: Two words are anagrams if they have the same letters but in a different order.)*

```python

def anagrm(wd1,wd2):
  return sorted(wd1)== sorted(wd2)


anagrm("listen","silentd")
```

```python
anagrm('cheese','butter')
```

*Write a function to check if a word is an isogram. (Definition: An isogram is a word or phrase with no repeating letters.)*

```python

def isogrm(word):
  word=word.lower()
  return len(set(word))== len(word)



isogrm('running')
```

```python
isogrm('import')
```

*Write a function to check if a word is a palindrome. (Definition: A palindrome is a word, number, phrase, or sequence that reads the same backward as forward.)*

```python
def palin(word):
  word=word.lower()
  wordrev=word[::-1]

  if word==wordrev:
    return True
  else:
    return False


palin('malayalam')
```

```python
palin('pig')
```
